```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
data = pd.read_csv('Iris.csv')
data.head()
```

```python
# Step 2: Load the Iris dataset
iris = load_iris()
iris_data = iris.data  # The features (sepal length, sepal width, petal length, petal width)
iris_feature_names = iris.feature_names  # Feature names
```

```python
df_iris = pd.DataFrame(data=iris_data, columns=iris_feature_names)
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df_iris.values, labels=df_iris.columns)
plt.title("Box Plot of Iris Dataset Features")
plt.xlabel("Features")
plt.ylabel("Values")
#plt.grid(True)
plt.show()
```

```python
#scatter plot
iris = load_iris()
iris_data = iris.data  # The features (sepal length, sepal width, petal length, petal width)
iris_feature_names = iris.feature_names  # Feature names
iris_target = iris.target  # Species as numerical labels
iris_target_names = iris.target_names  # Species names
```

```python
# Step 3: Convert to DataFrame for easier manipulation
df_iris = pd.DataFrame(data=iris_data, columns=iris_feature_names)
df_iris['species'] = iris_target
```

```python
# Step 4: Create Scatter Plot
plt.figure(figsize=(10, 6))

# Scatter plot of sepal length vs. sepal width
plt.scatter(df_iris['sepal length (cm)'], df_iris['sepal width (cm)'], c=df_iris['species'], cmap='viridis')
plt.title("Scatter Plot of Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.colorbar(label='Species')
plt.show()
```

```python
#bar plot
# Calculate the mean of each feature grouped by species
mean_features = df_iris.groupby('species').mean()
```

```python
plt.figure(figsize=(10, 6))

# Create a bar plot for each feature
mean_features.plot(kind='bar', ax=plt.gca())
plt.title("Bar Plot of Mean Features by Species")
plt.xlabel("Species")
plt.ylabel("Mean Values")
plt.xticks(ticks=np.arange(len(iris_target_names)), labels=iris_target_names, rotation=0)
#plt.grid(axis='y')

plt.show()
```
